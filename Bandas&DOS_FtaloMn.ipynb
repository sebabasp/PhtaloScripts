```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
```

```python
plt.rcParams["figure.dpi"]=150
plt.rcParams["figure.facecolor"]="white"
plt.rcParams["figure.figsize"]=(8, 6)

fig, axes = plt.subplots(nrows=1, ncols=2)
plt.subplot(1,2,1)
data = np.loadtxt('Mn.dat.gnu')

k = np.unique(data[:, 0])
bands = np.reshape(data[:, 1], (-1, len(k)))

for band in range(len(bands)):
    axes[0].plot(k, bands[band, :]+2.5394, linewidth=2, alpha=0.5, color='k')
plt.xlim(min(k), max(k))
plt.ylim(-2,2)
plt.axhline(0, linestyle=(0, (5, 5)), linewidth=0.75, color='k', alpha=0.5)
plt.axvline(0.7059, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.5079, linewidth=0.75, color='k', alpha=0.5)
plt.axvline(1.8887, linewidth=0.75, color='k', alpha=0.5)

plt.xticks(ticks= [0, 0.7080, 1.5122, 1.8936], \
           labels=['$\Gamma$', 'M', 'X', '$\Gamma$'], fontsize = 15)
plt.ylabel("E - E$_F$ (eV)", fontsize = 15)

def data_loader(fname):
    import numpy as np

    data = np.loadtxt(fname)
    energy = data[:, 0]
    uppdos = data[:, 1] 
    dwpdos = data[:, 2] 

    return energy, uppdos, dwpdos

energy, uppdos_C, dwpdos_C = data_loader('atom_C_tot.dat')
_, uppdos_H, dwpdos_H = data_loader('atom_H_tot.dat')
_, uppdos_N, dwpdos_N = data_loader('atom_N_tot.dat')
_, uppdos_Mn, dwpdos_Mn = data_loader('atom_Mn_tot.dat')
plt.subplot(1,2,2)
axes[1].plot( -dwpdos_C, energy+2.5394, linewidth=1, color='#006699', label='C')
axes[1].plot( uppdos_C, energy+2.5394, linewidth=1, color='#006699')
axes[1].plot( -dwpdos_H, energy+2.5394, linewidth=1, color='r', label='H')
axes[1].plot( uppdos_H, energy+2.5394, linewidth=1, color='r')
axes[1].plot( -dwpdos_N, energy+2.5394, linewidth=1, color='k', label='N')
axes[1].plot( uppdos_N, energy+2.5394, linewidth=1, color='k')
axes[1].plot( -dwpdos_Mn, energy+2.5394, linewidth=1, color='b', label='Mn')
axes[1].plot( uppdos_Mn, energy+2.5394, linewidth=1, color='b')
axes[1].axhline(y= 0, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.ylim(-2,2)
plt.xlim(-5,5)
axes[1].set_xticks([])
axes[1].set_yticks([])
axes[1].legend(frameon=False, fontsize = 15)

fig.tight_layout(pad=-0.69)
plt.show()
```
